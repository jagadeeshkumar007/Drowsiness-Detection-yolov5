```python
import uuid
import os 
import time
```

```python
image_path = os.path.join('yolov5\\data','images')
lables = ['awake','drowsy']
no_of_images = 11
```

```python
cap = cv2.VideoCapture(1)
for label in lables:
    print("collecting images for ",label)
    time.sleep(5)
    for i in range(no_of_images):
        print("collecting images for ",label,"image number ",i)
        ret,frame = cap.read()  
        img_name = os.path.join(image_path,label+'.'+str(uuid.uuid1())+'.jpg')
        print(img_name)
        cv2.imwrite(img_name,frame)
        cv2.imshow('Image collection',frame)
        time.sleep(2)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
        
```

```python
import tkinter as tk
import customtkinter as ctk

import torch
import numpy as np

import os
os.add_dll_directory("C:\Program Files (x86)\VideoLAN\VLC")

import cv2
from PIL import Image,ImageTk
import vlc

app = tk.Tk()
app.geometry("600x600")
app.title("Drowsy dio")
ctk.set_appearance_mode("dark")
vidFrame = tk.Frame(height=400,width=500)
vidFrame.pack()
vid = ctk.CTkLabel(vidFrame)
vid.pack()

counter = 0
counterLabel = ctk.CTkLabel(vidFrame,text=counter,height=40,width=120,font=("Arial",20),text_color="white",fg_color='teal')
counterLabel.pack(pady=10)

def reset_counter():
    global counter
    counter = 0

resetButton = ctk.CTkButton(vidFrame,text='Reset Counter',command=reset_counter,height=40,width=120,font=("Arial",20),text_color="white",fg_color='teal')
resetButton.pack()

model = torch.hub.load('ultralytics/yolov5','custom',path='yolov5/runs/train/exp5/weights/last.pt',force_reload=True)
cap = cv2.VideoCapture(1)

def detect():
    global counter
    ret,frame = cap.read()
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    results = model(frame)
    img = np.squeeze(results.render())
#     print(results.xywh)
    if len(results.xywh[0]) >0:
        dconf = results.xywh[0][0][4]
        dclass = results.xywh[0][0][5]
        
        if dconf.item() >0.65 and dclass.item() == 16.0:
            p = vlc.MediaPlayer("C:/Users/jagad/Downloads/clock-alarm-8761.mp3")
            p.play()
            counter+=1
            
    imgarr = Image.fromarray(img)
    imgtk  = ImageTk.PhotoImage(imgarr)
    vid.imgtk = imgtk
    vid.configure(image=imgtk)
    vid.after(10,detect)
    counterLabel.configure(text=counter)

detect()

app.mainloop()
```
